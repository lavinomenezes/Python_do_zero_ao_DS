```python
import ipywidgets as widgets
from ipywidgets import fixed
```

To make the filters with a graphic interface we will use the ipywidgets library

```python
data_mapa = df[['id','lat','long','price','level','waterfront','bathrooms','sqft_living','condition_type','sqft_basement']].copy()
data_mapa['yr_buid'] = pd.to_datetime(df['yr_built']).dt.strftime('%Y-%m')
data_mapa['waterfront'] = data_mapa['waterfront'].apply(lambda x: 'Yes' if x == 1 else 'No');
data_mapa.head()
```

```python
data_mapa['yr_buid'].value_counts()
```

```python
data_mapa['waterfront'].value_counts()
```

```python
style = {'description_width':'initial'}
waterfront_bar = widgets.ToggleButtons(
    options=['No','Yes'],
    description='Waterfront',
    disabled=False,
    button_style='', 
    tooltips=['yes','no'],
    style = style

)
condition_bar = widgets.ToggleButtons(
    options=['bad','regular','good'],
    description='Condition  ',
    disabled=False,
    button_style='', 
    tooltips=['bad','regular','good'],
    style = style

)
price_limit = widgets.IntRangeSlider(
    value=[int(data_mapa['price'].min()), int(data_mapa['price'].max())],
    min = int(data_mapa['price'].min()),
    max = int(data_mapa['price'].max()),
    step=1,
    description='Price range',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
   style = style
)
bathroom_limit = widgets.IntSlider(
    value = 0,
    min = data_mapa['bathrooms'].min(),
    max = data_mapa['bathrooms'].max(),
    step = 1,
    description = 'Bathrooms',
    disable=False,
    style = style
)
liv_limit = widgets.IntRangeSlider(
    value=[data_mapa['sqft_living'].min(), data_mapa['sqft_living'].max()],
    min = data_mapa['sqft_living'].min(),
    max = data_mapa['sqft_living'].max(),
    step=10,
    description='Living size',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
    style = style
)
base_limit = widgets.IntRangeSlider(
    value=[data_mapa['sqft_basement'].min(), data_mapa['sqft_basement'].max()],
    min = data_mapa['sqft_basement'].min(),
    max = data_mapa['sqft_basement'].max(),
    step=10,
    description='Basement size',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
    style = style
)


def updata_map(df,cond, waterfront,limit,bath,liv,base):
    houses = df[((df['price'] >= limit[0])&(df['price'] <= limit[1])) & (df['waterfront'] == waterfront) & (df['bathrooms'] <= bath) & ((df['sqft_living']>=liv[0])& (df['sqft_living']<=liv[1]))&(df['condition_type']==cond) & ((df['sqft_basement']>=base[0])& (df['sqft_basement']<=base[1]))].copy()
    print("There are:",len(houses),"houses available with this conditions")
    fig = px.scatter_mapbox(houses,lat="lat",lon="long",color='level',size='price',hover_name="id",hover_data=["price","bathrooms","sqft_living","sqft_basement"],size_max=15,color_continuous_scale=px.colors.cyclical.IceFire,zoom=8,height=800)
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
    
```

```python
widgets.interactive(updata_map,df = fixed(data_mapa),cond = condition_bar,waterfront = waterfront_bar,limit = price_limit,bath = bathroom_limit,liv=liv_limit,base = base_limit)
```

### 5. Add filter's options on the last dashboard 
**- Visualize values from a certain date;** 

**- Filter by year of renovation;**

**- Filter properties with waterfront or not.**



```python
df_dash = df.copy()
```

copying the original dataframe to be able to do some transformations without changing the data

```python
df_dash['year'] = pd.to_datetime(df_dash['date']).dt.strftime('%y')
df_dash['date'] = pd.to_datetime(df_dash['date']).dt.strftime('%Y-%m-%d')
df_dash['yr_renovated'] = pd.to_datetime(df_dash['yr_renovated']).dt.year
df_dash['year_week'] = pd.to_datetime(df_dash['date']).dt.strftime('%Y-%U')
```

```python
df_dash.head()
```

```python
date_limit = widgets.SelectionSlider(
    options = df_dash['date'].sort_values().unique().tolist(),
    value = '2014-12-01',
    description = 'Disponivel',
    continuous_update = False,
    orientation = 'horizontal',
    readout= True
)
year_limit = widgets.SelectionSlider(
    options = df_dash['yr_renovated'].sort_values().unique().tolist(),
    value = 2000,
    description = 'Max year',
    continuous_update = False,
    orientation = 'horizontal',
    readout= True
)
waterfront_bar = widgets.ToggleButtons(
    options=['no','yes'],
    description='Waterfront',
    disabled=False,
    button_style='', 
    tooltips=['no','yes'],

)
def update_dash(data,limit,year,front):
    #filter data
    data['waterfront'] = data['waterfront'].apply(lambda x: 'yes' if x == 1 else 'no')
    df = data[(data['date'] <=limit)&(data['yr_renovated']>=year)&(data['waterfront']==front)].copy()
    
    fig = plt.figure(figsize=(21,12))
    specs= gridspec.GridSpec(ncols=2,nrows=2,figure = fig)
    
    ax1 = fig.add_subplot(specs[0,:])
    ax2 = fig.add_subplot(specs[1,0])
    ax3 = fig.add_subplot(specs[1,1])
    
    by_year = df[['price','year']].groupby('year').sum().reset_index()
    ax1.bar(by_year['year'],by_year['price'])
    set_currency(ax1)
    ax1.set_title('Avg Price by Year')
    
    by_day = df[['price','date']].groupby('date').mean().reset_index()
    ax2.plot(by_day['date'],by_day['price'])
    ax2.set_xticklabels(by_day['date'].unique().tolist(),rotation=60)
    set_currency(ax2)
    ax2.set_title('Avg Price by Day')

    
    by_year_week = df[['price','year_week']].groupby('year_week').mean().reset_index()
    ax3.bar(by_year_week['year_week'],by_year_week['price'])
    set_currency(ax3)
    ax3.set_title('Avg Price by Year Week')
    plt.xticks(rotation=60);
```

```python
widgets.interact(update_dash,data = fixed(df_dash),limit = date_limit,year = year_limit,front = waterfront_bar);
```
